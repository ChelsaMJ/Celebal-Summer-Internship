### Import Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Line Graph

```python
sales_data = pd.read_csv('sales_data_50.csv')
monthly_sales_by_region = sales_data.groupby(['Month', 'Region'])['Sales'].sum().unstack()
month_order = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 
               'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
monthly_sales_by_region = monthly_sales_by_region.reindex(month_order)

plt.figure(figsize=(10,6))
plt.title('Monthly Sales by Region', fontdict={'fontweight':'bold', 'fontsize': 18})

for region in monthly_sales_by_region.columns:
    plt.plot(monthly_sales_by_region.index, 
             monthly_sales_by_region[region], 
             marker='o', label=region)

plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig('Monthly_Sales_by_Region.png', dpi=300)
plt.show()
```

### Histogram

```python
employee_data = pd.read_csv('employee_data.csv')

bins = [20, 25, 30, 35, 40, 45, 50]

plt.figure(figsize=(8,5))
plt.hist(employee_data.Age.dropna(), bins=bins, color='#abcdef')
plt.xticks(bins)
plt.ylabel('Number of Employees')
plt.xlabel('Age')
plt.title('Distribution of Employee Ages')
plt.savefig('age_histogram.png', dpi=300)
plt.show()
```

### Pie chart

```python
employee_data = pd.read_csv('employee_data.csv')

dept_counts = employee_data['Department'].value_counts()

plt.figure(figsize=(8,5))
plt.pie(dept_counts, labels=dept_counts.index, autopct='%.2f %%', colors=['#abcdef', '#aabbcc', '#99ddff', '#77ccdd'])
plt.title('Distribution of Employees by Department')
plt.show()
```

### Pie chart # 2

```python
sales_data = pd.read_csv('sales_data_50.csv')

region_counts = sales_data['Region'].value_counts()

plt.figure(figsize=(8, 5), dpi=100)
plt.style.use('ggplot')

explode = [0.05] * len(region_counts)

plt.pie(region_counts, 
        labels=region_counts.index, 
        autopct='%.2f %%', 
        explode=explode, 
        pctdistance=0.8)

plt.title('Sales Distribution by Region')
plt.show()
```

### Boxplot and Whiskers chart

```python
sales_data = pd.read_csv('sales_data_50.csv')

plt.figure(figsize=(5,8), dpi=100)
plt.style.use('default')

east = sales_data[sales_data.Region == 'East']['Sales']
west = sales_data[sales_data.Region == 'West']['Sales']
central = sales_data[sales_data.Region == 'Central']['Sales']

bp = plt.boxplot([east, west, central],
                 labels=['East', 'West', 'Central'],
                 patch_artist=True,
                 medianprops={'linewidth': 2})

plt.title('Sales Distribution by Region')
plt.ylabel('Sales Amount')

for box in bp['boxes']:
    box.set(color='#4286f4', linewidth=2)
    box.set(facecolor='#e0e0e0')

plt.show()
```
